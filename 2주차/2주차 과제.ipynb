사이킷런(sklearn)은 파이썬에서 머신러닝 분석을 할 때 유용하게 사용할 수 있는 라이브러리

sklearn.model_selection.train_test_split(*arrays, test_size=None, train_size=None, random_state=None, shuffle=True, stratify=None)

*arrays : sequence of indexables with same length / shape[0]  
분할시킬 데이터를 입력받습니다.  
list, numpy arrays, scipy-sparse matrices, pandas dataframes등이 들어올 수 있습니다.  

test_size : float or int, default=None  
테스트 세트의 구성 비율입니다.  
train_size의 옵션과 반대 관계에 있는 옵션 값이며, 주로 test_size를 지정해 줍니다.  
float의 경우 0.0 ~ 1.0의 값을 가지며 0.2라면 전체 데이터 세트의 20%를   test(validation) 세트로 지정하겠다는 의미입니다.  
int의 경우 테스트 세트의 크기를 정확한 값으로 표현합니다.  
None의 경우 1 - train_size의 값을 갖습니다.   
default 값은 0.25 입니다.  

train_size : float or int, default=None  
트레인 세트의 구성 비율입니다.  
test_size의 옵션과 반대 관계에 있는 옵션 값입니다.  
float의 경우 0.0 ~ 1.0의 값을 가지며 0.2라면 전체 데이터 세트의 20%를 train 세트로 지정하겠다는 의미입니다.  
int의 경우 트레인 세트의 크기를 정확한 값으로 표현합니다.  
None의 경우 1 - test_size의 값을 갖습니다.   
default 값은 0.75 입니다.  

random_state : int, RandomState instance or None, default=None  
세트를 섞을 때 해당 int 값을 보고 섞으며, 하이퍼 파라미터를 튜닝시 이 값을 고정해두고 튜닝해야 매번 데이터셋이 변경되는 것을 방지할 수 있습니다.  

shuffle : bool, default=True  
split하기 전에 데이터를 섞을지 여부를 설정합니다.  

stratify : array-like, default=None  
지정한 Data의 비율을 유지합니다. 예를 들어, Label Set인 Y가 25%의 0과 75%의 1로 이루어진 Binary Set일 때, stratify=Y로 설정하면 나누어진 데이터셋들도 0과 1을 각각 25%, 75%로 유지한 채 분할됩니다.

plt.subplots를 통해서 3가지 1행 3열의 subplots를 생성합니다.  
이 때 axes는 3개의 plot객체가 저장됩니다.  

mglearn.plots.plot_2d_separator는 모델 객체(clf)와 train Data(X)를 주고 데이터세트 분할 평면을 그리는 함수입니다. 이 때 파라미터에 있는 eps는 epsilon의 약자로 SVM에서 사용된다고 하는데 아직은 잘 모르겠습니다.

```python
fig, axes = plt.subplots(1, 3, figsize=(10,3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
axes[0].legend(loc=3)
```

아래의 코드는 K값을 1~11까지 바꿔가며 이 때의 정확도(clf.score(X_,y_)를 training_accuracy와 test_accuracy에 넣고 그 값의 변화를 그래프로 나타내는 코드입니다.

```python
training_accuracy = []
test_accuracy = []
neighbors_settings = range(1,11)
```

```python
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
```

```python
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

아래의 코드는 위에서 cancer에 대한 코드를 보고 "아래 예시도 똑같이 적용되지 않을까?"라고 생각하여 실행한 코드입니다. 에러가 뜬 이유는 boston객체에는 target_names라는 키가 없었습니다. 또한, boston.target은 float값들로 구성된 배열인데 bincount에는 int값들로 구성된 배열만 넣을 수 있으므로 이 부분도 에러가 발생합니다.

아래의 코드는 "X_train의 size가 19인데 과연 K값이 이보다 커지면 어떻게 될까?"라는 의문에서 작성해보았고 K값이 size보다 커지면 에러가 발생하고 실행되지 않음을 알 수 있었습니다.

```python
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

training_accuracy = []
test_accuracy = []
neighbors_settings = range(1,50)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

아래의 코드는 유방암 데이터에 대해서 "같은 data에 대해서 random_state값을 바꾸었을 때 정확도에 영향을 미치는가?"라는 질문을 가지고 작성해보았습니다. 그 결과로 그래프가 복잡하고 확실하지는 않지만 그 정확도가 변화한다는 사실을 확인할 수 있었습니다.
